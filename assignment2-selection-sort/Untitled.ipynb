```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "ArraySize": [100,100,100,500,500,500,1000,1000,1000,5000,5000,5000,
                  10000,10000,10000,20000,20000,20000,50000,50000,50000,
                  100000,100000,100000],
    "DataType": ["Random","Sorted","Reversed"]*8,
    "Time_ms": [0.4952,0.0167,0.2529,3.8383,0.0390,0.7975,3.9184,0.0609,3.0938,
                46.5933,0.0516,29.9693,160.2972,0.0784,119.9848,644.1011,0.1578,480.9050,
                4003.5668,0.3948,3001.8234,16084.6328,0.7773,12015.2970],
    "Comparisons": [4952,167,3823,124760,998,94123,499510,1998,375748,
                    12497507,9998,9378748,49995014,19998,37507498,199990035,39998,150014998,
                    1249975036,99998,937537498,4999950023,199998,3750074998],
    "Swaps": [4956,198,3823,495,0,250,994,0,500,4996,0,2500,9994,0,5000,19985,0,10000,
              49985,0,25000,99990,0,50000],
    "ArrayAccesses": [10300,396,7846,251500,1996,189246,1002996,3996,753496,
                      25014998,19996,18767496,100030004,39996,75034996,400060010,79996,300069996,
                      2500150012,199996,1875174996,10000300006,399996,7500349996]
}

df = pd.DataFrame(data)

metrics = ["Time_ms", "Comparisons", "Swaps", "ArrayAccesses"]
y_labels = ["Time (ms)", "Comparisons", "Swaps", "Array Accesses"]

#  
for metric, y_label in zip(metrics, y_labels):
    plt.figure(figsize=(10,6))
    for dtype in df['DataType'].unique():
        subset = df[df['DataType']==dtype]
        plt.plot(subset['ArraySize'], subset[metric], marker='o', label=dtype)
    plt.title(f'SelectionSort {y_label} Benchmark')
    plt.xlabel('Array Size')
    plt.ylabel(y_label)
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
